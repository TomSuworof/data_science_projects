## Интерфейс (6 баллов)
Алгоритм кластеризации должен удовлетворять следующему интерфейсу. Конструктор принимает набор параметров, необходимых для работы алгоритма кластеризации. Метод `fit` подсчитывает параметры модели и возвращает `self`. Метод `predict` возвращает вектор с индексами кластеров для поданных в него объектов `x`

```python
from sklearn.base import BaseEstimator, ClusterMixin
from queue import Queue

class Clustering(BaseEstimator, ClusterMixin):
    """
    Implement clustering algorithm  
    """
    
    def __init__(self,
                 eps: float = 0.5,
                 min_samples: int = 5,
                 **kwargs):
        """
        Please add necessary algoritm parameters to class constructor.
        """
        self.eps = eps
        self.min_samples = min_samples
        
        # введём некоторые обозначения
        self.noise = -1  # шумы
        self.unassigned = 0  # неопределённые (таких не будет по окончанию алгоритма)
        self.core = 0  # опорные
        self.reachable = -2  # достижимые или граничные
        
        self.core_sample_indices_ = []  # [n_core_samples]
        self.components_ = []  # [n_core_samples, n_features]
        self.labels_ = []  # [n_samples]
        
        self.kneighbors = {}
        
        
    def euclidean_dist(self, p, q):
        # обычный алгоритм нахождения евклидова расстояния
        # только для числовых признаков - надо отсеивать их
        return sum([(x - y)**2 for x, y in zip(p, q)])**(1/2)

    
    def nearest_neighbors(self, point_id, radius):
        neighbors = []
        for i in range(self.x_len):
            if self.euclidean_dist(self.components_[i], self.components_[point_id]) <= radius:  # если расстояние меньше заданного
                neighbors.append(i)  # то это сосед, запишем его
        return neighbors

        
    def fit(self, x, y=None):
        """
        Use data matrix x to compute model parameters
        """
        self.x_len = len(x)
        self.components_ = pd.DataFrame(x).to_numpy()  # избавимся от разносортности данных
        
        # этот список будет хранить статусы точек
        self.labels_ = [self.unassigned] * self.x_len
        
        # запишем ближайших соседей для каждой точки
        self.point_count = []
        for i in range(self.x_len):
            self.point_count.append(self.nearest_neighbors(i, self.eps))
            
        # списки опорных и остальных (core and reachable + noise) точек
        self.core_sample_indices_ = []
        self.noncore_sample_indices_ = []
                
        for i in range(len(self.point_count)):
            if len(self.point_count[i]) >= self.min_samples:  # если у точки нужное число соседей (включая себя)
                self.labels_[i] = self.core  # то точка - опорная
                self.core_sample_indices_.append(i)  # записали эту точку в список опорных
            else:
                self.noncore_sample_indices_.append(i)  # иначе это не-опорная точка
                
        # пройдёмся по не-опрным точкам
        for i in self.noncore_sample_indices_:
            for j in self.point_count[i]: # пройдёмся по соседям не-опорной точки
                if j in self.core_sample_indices_:  # если у i-ой точки соседняя j-ая точка является опорной
                    self.labels_[i] = self.reachable  # то она (i-ая) считается достижимой
                    break  # достаточно одного опорного сосдеа, чтобы считатться достижимой
                        
        return self
    
    def predict(self, x):
        """
        Using computed model parameters predict cluster
        for all objects from x
        """
        
        # будем использовать очередь для хранения соседних опорных точек для нахождения их соседей
        cluster = 1
                
        for i in range(len(self.labels_)):
            q = Queue()
            if self.labels_[i] == self.core:  # если точка опорная
                self.labels_[i] = cluster  # то запишем её в текущий кластер
                for pt_idx in self.point_count[i]:  # просмотрим соседние точки                    
                    if self.labels_[pt_idx] == self.core:  # если соседняя точка опорная
                        q.put(pt_idx)  # то запомним её (потом понадобится)
                        self.labels_[pt_idx] = cluster  # и занесём в текущий кластер
                    elif self.labels_[pt_idx] == self.reachable:
                        self.labels_[pt_idx] = cluster  # достижимые точки заносим в кластер, но не сохраняем в очередь
                # теперь пройдём по сохранённым основным точкам-соседям
                while not q.empty():
                    neighbors = self.point_count[q.get()]  # найдём соседей точки
                    # и повторим то же самое, что и раньше
                    for pt_idx in neighbors:
                        if self.labels_[pt_idx] == self.core:
                            q.put(pt_idx)
                            self.labels_[pt_idx] = cluster
                        elif self.labels_[pt_idx] == self.reachable:
                            self.labels_[pt_idx] = cluster
                    # опорная точка -> её опорные соседи. и так для каждой опорной точки
                cluster += 1  # переход к следующему кластеру, когда закончим с опорными соседями
        return self.labels_
    
    def fit_predict(self, x, y=None):
#         self.fit(x, y)
        return self.fit(x, y).predict(x)
```

Своя реализация NearestNeighbors:

```python
def euclidean_dist(p, q):
    return sum([(x - y)**2 for x, y in zip(p, q)])**(1/2)


def kneighbors(n_neighbors, data):
    indices = []
    dists = []
    for i in range(len(data)):
        idxs_dists = {}  # сюда будем записывать индексы и расстояния до точек
        
        for j in range(len(data)):
            idxs_dists[j] = euclidean_dist(data[i], data[j])  # собрали данные о других точках
        top_idxs_dists = sorted(idxs_dists.items(), key=lambda x: x[1])[0:n_neighbors]  # отобрали ближайших
        
        indices.append([x[0] for x in top_idxs_dists])
        dists.append([x[1] for x in top_idxs_dists])
        
    return np.array(dists), np.array(indices)
```

```python
from sklearn.neighbors import NearestNeighbors

for i in range(3, 4):
    nbrs = NearestNeighbors(n_neighbors=i).fit(x_iris)
    distances, indices = nbrs.kneighbors(x_iris)
    
#     print(distances)
#     print(indices)

    plt.plot(sorted(distances[:, -1]))
    plt.show()
```

```python
for i in range(3, 10):
    distances, indices = kneighbors(i, x_iris)
    
#     print(distances)
#     print(indices)
    print('Clusters number:', i)
    plt.plot(sorted(distances[:, -1]))
    plt.show()
```

### Применение алгоритма (2 балла)

```python
# http://scikit-learn.org/stable/auto_examples/cluster/plot_cluster_comparison.html
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from time import time
from sklearn import datasets
from sklearn.preprocessing import StandardScaler

np.random.seed(0)

n_samples = 1500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5, noise=.05)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
no_structure = np.random.rand(n_samples, 2), None

data = [noisy_circles, noisy_moons, blobs, no_structure]
```

```python
plt.figure(figsize=(2 + 3, 9.5))
for i_dataset, dataset in enumerate(data):
    X, y = dataset
    # normalize dataset for easier parameter selection
    X = StandardScaler().fit_transform(X)
    c = [0] * len(X)
    # plot
    plt.subplot(4, 1, i_dataset + 1)
    plt.scatter(X[:, 0], X[:, 1], s=10, c=c)
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.xticks(())
    plt.yticks(())

plt.show()
```

```python
eps, min_samples = 0.3, 5
```

```python
my_scores = []
db_scores = []

my_times = []
db_times = []
```

```python
plt.figure(figsize=(2 + 3, 9.5))
for i_dataset, dataset in enumerate(data):
    X, y = dataset
    # normalize dataset for easier parameter selection
    X = StandardScaler().fit_transform(X)
    t = time()
    cl = Clustering(eps=eps, min_samples=min_samples).fit_predict(X)
    t = time() - t
    my_times.append(t)
    try:
        my_scores.append(smt.adjusted_rand_score(y, cl))
    except TypeError:
        pass   
    # plot
    plt.subplot(4, 1, i_dataset + 1)
    plt.scatter(X[:, 0], X[:, 1], s=10, c=cl)
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.xticks(())
    plt.yticks(())

plt.show()
```

Теперь примените готовый алгоритм из библиотеки sklearn на тех же данных. 
Сравните результат кластеризации и время, за которое отрабатывает ваш алгоритм и алгоритм из sklearn.

```python
from sklearn.cluster import DBSCAN

plt.figure(figsize=(2 + 3, 9.5))
for i_dataset, dataset in enumerate(data):
    X, y = dataset
    # normalize dataset for easier parameter selection
    X = StandardScaler().fit_transform(X)
    t = time()
    cl = DBSCAN(eps=0.3, min_samples=5).fit_predict(X)
    t = time() - t
    db_times.append(t)
    try:
        db_scores.append(smt.adjusted_rand_score(y, cl))
    except TypeError:
        pass
    # plot
    plt.subplot(4, 1, i_dataset + 1)
    plt.scatter(X[:, 0], X[:, 1], s=10, c=cl)
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.xticks(())
    plt.yticks(())

plt.show()
```

```python
for x, y in zip(my_scores, db_scores):
    print(x, y)
```
