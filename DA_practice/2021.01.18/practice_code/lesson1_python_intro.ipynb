### Классы

```python
class Greeter:

    # Constructor
    def __init__(self, name):
        self.name = name

    # Instance method
    def greet(self, loud=False):
        if loud:
            print(f'HELLO, {self.name.upper()}')
        else:
            print('Hello, {}'.format(self.name))
```

```python
g = Greeter('Fred')
g
```

```python
g.greet()
```

```python
g.greet(True)
```

### Функции

```python
def sign(x):
    if x > 0:
        return 'positive'
    elif x < 0:
        return 'negative'
    else:
        return 'zero'

for x in [-1, 0, 1]:
    print(sign(x))
```

## Математические операции с массивами

```python
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)
print(x)
print()
print(y)
```

```python
print(x + y)
```

```python
print(x - y)
```

```python
print(x * y)
```

```python
print(x * 2)
```

```python
print(x @ y)
print(x.dot(y))
print(np.dot(x, y))
```

```python
print(np.sum(x))
print(np.sum(x, axis=0))
print(np.sum(x, axis=1))
```
