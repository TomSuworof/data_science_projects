```python
import gensim
import urllib
import collections
import os

import zipfile

import numpy as np
import tensorflow as tf
#import textacy
import spacy
```

```python
data = []
with open('./data/Rowling_J.__Harry_Potter_and_the_Sorcerers_Stone_www.Readli.Net_81716.txt','r') as file:
    for line in file:
        add = gensim.utils.simple_preprocess(line)
        if len(add)!=0:
            data.append(add)
```

```python
data
```

# Gensim 

```python
%%time
model = gensim.models.FastText(size=2, window=5, min_count=20)  
model.build_vocab(sentences=data)
model.train(sentences=data, total_examples=len(data), epochs=100)
```

```python
model.wv.vocab
```

```python
len(model.wv.vocab)
```

```python
model.wv['weasley']
```

```python
heros = ['harry', 'potter','weasley', 'hermione', 'neville', 'hagrid', 'dumbledore', 'muggle', 'ron',
        'mcgonagall', 'draco', 'malfoy', 'quirrell', 'severus', 'snape', 'professor', 'student', 'dursley']
```

```python
import time
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for k, i in enumerate(heros):
    #print(i)
    arr = model.wv[i]
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.text(arr[0], arr[1], i, fontsize=9)
    #time.sleep(1)
    fig.show()
    fig.canvas.draw()
```

## train model

```python
%%time
model = gensim.models.Word2Vec(size=100, alpha=0.025, window=5, min_count=20,
                               sample=0.001, seed=1, workers=3, min_alpha=0.0001, sg=1, hs=0, negative=5, iter=100, 
                               sorted_vocab=1, batch_words=10000, compute_loss=False)  
model.build_vocab(sentences=data)
model.train(sentences=data, total_examples=len(data), epochs=100)
```

```python
len(model.wv.vocab)
```

```python
model.wv['harry']
```

```python
model.wv.save('./data/our_wv.wv')
```

```python
model.wv.save_word2vec_format('./data/our_wv_voc.wv')
```
