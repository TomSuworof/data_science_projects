And finally, let's use the vocabulary you've built to map text lines into neural network-digestible matrices.

```python
UNK_IX, PAD_IX = map(token_to_id.get, [UNK, PAD])

def as_matrix(sequences, max_len=None):
    """ Convert a list of tokens into a matrix with padding """
    if isinstance(sequences[0], str):
        sequences = list(map(str.split, sequences))
        
    max_len = min(max(map(len, sequences)), max_len or float('inf'))
    
    matrix = np.full((len(sequences), max_len), np.int32(PAD_IX))
    for i,seq in enumerate(sequences):
        row_ix = [token_to_id.get(word, UNK_IX) for word in seq[:max_len]]
        matrix[i, :len(row_ix)] = row_ix
    
    return matrix
```

```python
import torch

def to_tensors(batch, device):
    batch_tensors = dict()
    for key, arr in batch.items():
        if key in ["FullDescription", "Title"]:
            batch_tensors[key] = torch.tensor(arr, device=device, dtype=torch.int64)
        else:
            batch_tensors[key] = torch.tensor(arr, device=device)
    return batch_tensors

def make_batch(data, max_len=None, word_dropout=0, device=torch.device('cpu')):
    """
    Creates a keras-friendly dict from the batch data.
    :param word_dropout: replaces token index with UNK_IX with this probability
    :returns: a dict with {'title' : int64[batch, title_max_len]
    """
    batch = {}
    batch['Id'] = data['Id'].values
    batch["Title"] = as_matrix(data["Title"].values, max_len)
    batch["FullDescription"] = as_matrix(data["FullDescription"].values, max_len)
    batch['Categorical'] = categorical_vectorizer.transform(data[categorical_columns].apply(dict, axis=1))
    
    if word_dropout != 0:
        batch["FullDescription"] = apply_word_dropout(batch["FullDescription"], 1. - word_dropout)
    
    if TARGET_COLUMN in data.columns:
        batch[TARGET_COLUMN] = data[TARGET_COLUMN].values
    
    return to_tensors(batch, device)

def apply_word_dropout(matrix, keep_prop, replace_with=UNK_IX, pad_ix=PAD_IX,):
    dropout_mask = np.random.choice(2, np.shape(matrix), p=[keep_prop, 1 - keep_prop])
    dropout_mask &= matrix != pad_ix
    return np.choose(dropout_mask, [matrix, np.full_like(matrix, replace_with)])
```

# Base model

#### Architecture

Our basic model consists of three branches:
* Title encoder
* Description encoder
* Categorical features encoder

We will then feed all 3 branches into one common network that predicts salary.

![scheme](https://github.com/yandexdataschool/nlp_course/raw/master/resources/w2_conv_arch.png)

This clearly doesn't fit into keras' __Sequential__ interface. To build such a network, one will have to use PyTorch.

```python
import torch
import torch.nn as nn
import torch.functional as F
```

```python
models_and_metrics = {}
```

```python
class SalaryPredictor(nn.Module):
    def __init__(self, n_tokens=len(tokens), n_cat_features=len(categorical_vectorizer.vocabulary_), hid_size=8):
        super().__init__()
        self.embedder = nn.Embedding(n_tokens, hid_size)
        self.title_encoder = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=2),
            nn.Dropout(p=0.25),
            nn.ReLU(),
            nn.AdaptiveMaxPool1d(output_size=1)
        )
        self.description_encoder = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=2),
            nn.Dropout(p=0.25),
            nn.ReLU(),
            nn.AdaptiveMaxPool1d(output_size=1)
        )
        self.categorical_encoder = nn.Sequential(
            nn.Linear(n_cat_features, hid_size * 2),
            nn.ReLU(),
            nn.Linear(hid_size * 2, hid_size * 2),
            nn.ReLU()
        )
        self.final_predictor = nn.Sequential(
            nn.Linear(hid_size * 4, hid_size),
            nn.ReLU(),
            nn.Linear(hid_size, 1)
        )

        
    def forward(self, batch):
        title_embeddings = self.embedder(batch['Title']).permute(0, 2, 1)
        title_features = self.title_encoder(title_embeddings).squeeze()

        description_embeddings = self.embedder(batch['FullDescription']).permute(0, 2, 1)
        description_features = self.description_encoder(description_embeddings).squeeze()

        categorical_features = self.categorical_encoder(batch['Categorical'])

        features = torch.cat(
            [title_features, description_features, categorical_features], dim=1)
        
        return self.final_predictor(features).squeeze()
```

#### Training and evaluation

As usual, we gonna feed our monster with random minibatches of data. 

As we train, we want to monitor not only loss function, which is computed in log-space, but also the actual error measured in dollars.

```python
def iterate_minibatches(data, batch_size=256, shuffle=True, cycle=False, **kwargs):
    """ iterates minibatches of data in random order """
    while True:
        indices = np.arange(len(data))
        if shuffle:
            indices = np.random.permutation(indices)

        for start in range(0, len(indices), batch_size):
            batch = make_batch(data.iloc[indices[start : start + batch_size]], **kwargs)
            yield batch
        
        if not cycle: break
```

### Model training

We can now fit our model the usual minibatch way. The interesting part is that we train on an infinite stream of minibatches, produced by `iterate_minibatches` function.

```python
import tqdm
```

```python
def return_metrics(model, data, batch_size, name="", **kw):
    squared_error = abs_error = num_samples = 0.0
    model.eval()
    with torch.no_grad():
        for batch in iterate_minibatches(data, batch_size=batch_size, shuffle=False, **kw):
            batch_pred = model(batch)
            squared_error += torch.mean(torch.square(batch_pred - batch[TARGET_COLUMN]))
            abs_error += torch.mean(torch.abs(batch_pred - batch[TARGET_COLUMN]))
            num_samples += len(batch)
    mse = squared_error.detach().cpu().numpy() / num_samples
    mae = abs_error.detach().cpu().numpy() / num_samples
    return mse, mae
```

```python
BATCH_SIZE = 512
EPOCHS = 3
DEVICE = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

model_base = SalaryPredictor().to(DEVICE)
criterion = nn.MSELoss(reduction='mean')
optimizer = torch.optim.Adam(model_base.parameters(), lr=1e-3)

losses, mse, mae = [], [], []

for epoch in range(EPOCHS):
    print(f"epoch: {epoch}")
    for i, batch in tqdm.notebook.tqdm(enumerate(
            iterate_minibatches(
                    data_train, 
                    batch_size=BATCH_SIZE, 
                    device=DEVICE)), 
                total=len(data_train) // BATCH_SIZE
        ):
        model_base.train()
        pred = model_base(batch)
        optimizer.zero_grad()
        loss = criterion(pred, batch[TARGET_COLUMN])
        loss.backward()
        optimizer.step()

        if i % 100 == 99:
            losses.append(loss.item())
            mse_item, mae_item = return_metrics(model_base, data_val, name='val', batch_size=BATCH_SIZE, device=DEVICE)
            mse.append(mse_item)
            mae.append(mae_item)

fig, ax = plt.subplots(nrows=3, ncols=1)

ax[0].plot(losses, label=f'loss, end={losses[-1]}')
ax[1].plot(mse, label=f'mse, end={mse[-1]}')
ax[2].plot(mae, label=f'mae, end={mae[-1]}')

for i in range(3):
    ax[i].legend(loc='upper right')

plt.show()
```

```python
models_and_metrics['base'] = {}
models_and_metrics['base']['losses'] = losses
models_and_metrics['base']['mse'] = mse
models_and_metrics['base']['mae'] = mae
```

```python
class SalaryPredictor3(nn.Module):
    def __init__(self, n_tokens=len(tokens), n_cat_features=len(categorical_vectorizer.vocabulary_), hid_size=8):
        super().__init__()
        self.embedder = nn.Embedding(n_tokens, hid_size)
        self.title_encoder_head1 = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=2)
        )
        self.title_encoder_head2 = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=2)
        )
        self.title_encoder_tail = nn.Sequential(
            nn.BatchNorm1d(hid_size * 2),
            nn.Dropout(p=0.25),
            nn.ReLU(),
            nn.AdaptiveMaxPool1d(output_size=1),
            nn.ReLU()
        )
        self.description_encoder_head1 = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=2)
        )
        self.description_encoder_head2 = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=2)
        )
        self.description_encoder_tail = nn.Sequential(
            nn.BatchNorm1d(hid_size * 2),
            nn.Dropout(p=0.25),
            nn.ReLU(),
            nn.AdaptiveMaxPool1d(output_size=1),
            nn.ReLU()
        )
        self.categorical_encoder = nn.Sequential(
            nn.Linear(n_cat_features, hid_size * 2),
            nn.ReLU(),
            nn.Linear(hid_size * 2, hid_size * 2),
            nn.ReLU()
        )
        self.final_predictor = nn.Sequential(
            nn.Linear(hid_size * 6, hid_size),
            nn.ReLU(),
            nn.Linear(hid_size, 1),
            nn.ReLU()
        )

        
    def forward(self, batch):
        title_embeddings = self.embedder(batch['Title']).permute(0, 2, 1)

        title_features1 = self.title_encoder_head1(title_embeddings).squeeze()
        title_features2 = self.title_encoder_head2(title_embeddings).squeeze()

        title_features = torch.cat([title_features1, title_features2], dim=1)
        title_features = self.title_encoder_tail(title_features).squeeze()

        description_embeddings = self.embedder(batch['FullDescription']).permute(0, 2, 1)

        description_features1 = self.description_encoder_head1(description_embeddings).squeeze()
        description_features2 = self.description_encoder_head2(description_embeddings).squeeze()

        description_features = torch.cat([description_features1, description_features2], dim=1)
        description_features = self.description_encoder_tail(description_features).squeeze()

        categorical_features = self.categorical_encoder(batch['Categorical'])

        features = torch.cat(
            [
             title_features,
             description_features,
             categorical_features
             ], dim=1)
        return self.final_predictor(features).squeeze()
```

```python
early_stopping_patience = 5
early_stopping_tol = 0.1

BATCH_SIZE = 512
EPOCHS = 3
DEVICE = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

model5 = SalaryPredictor4().to(DEVICE)
criterion = nn.MSELoss(reduction='mean')
optimizer = torch.optim.Adam(model5.parameters(), lr=1e-3)

losses, mse, mae = [], [], []

patience_passed = 0
plateau_loss = 999

for epoch in range(EPOCHS):
    print(f"epoch: {epoch}")
    for i, batch in tqdm.notebook.tqdm(enumerate(
            iterate_minibatches(
                    data_train, 
                    batch_size=BATCH_SIZE, 
                    device=DEVICE)), 
                total=len(data_train) // BATCH_SIZE
        ):
        model5.train()
        pred = model5(batch)
        optimizer.zero_grad()
        loss = criterion(pred, batch[TARGET_COLUMN])
        loss.backward()
        optimizer.step()

        if plateau_loss - early_stopping_tol <= loss.item() <= plateau_loss: # если между плато-толерантность и плато, то ждём
            patience_passed += 1
        elif loss.item() < plateau_loss - early_stopping_tol: # если сильно улучшили, то обнуляем ожидание
            plateau_loss = loss.item()
            patience_passed = 0

        if patience_passed > early_stopping_patience: # хватит ждать
            break

        if i % 100 == 99:
            losses.append(loss.item())
            mse_item, mae_item = return_metrics(model5, data_val, name='val', batch_size=BATCH_SIZE, device=DEVICE)
            mse.append(mse_item)
            mae.append(mae_item)
    
    # helps to break outer loop
    else:
        continue
    break

fig, ax = plt.subplots(nrows=3, ncols=1)

ax[0].plot(losses, label=f'loss, end={losses[-1]}')
ax[1].plot(mse, label=f'mse, end={mse[-1]}')
ax[2].plot(mae, label=f'mae, end={mae[-1]}')

for i in range(3):
    ax[i].legend(loc='upper right')

plt.show()
```

```python
def as_matrix_gensim(sequences, max_len=None):
    """ Convert a list of tokens into a matrix with padding """
    if isinstance(sequences[0], str):
        sequences = list(map(str.split, sequences))
        
    max_len = min(max(map(len, sequences)), max_len or float('inf'))
    
    matrix = np.full((len(sequences), max_len), np.int32(PAD_IX))
    for i, seq in enumerate(sequences):
        row_ix = [vocab_to_id.get(word, UNK_IX) for word in seq[:max_len]]
        matrix[i, :len(row_ix)] = row_ix
    
    return matrix
```

```python
def make_batch_gensim(data, max_len=None, word_dropout=0, device=torch.device('cpu')):
    """
    Creates a keras-friendly dict from the batch data.
    :param word_dropout: replaces token index with UNK_IX with this probability
    :returns: a dict with {'title' : int64[batch, title_max_len]
    """
    batch = {}
    batch['Id'] = data['Id'].values
    batch["Title"] = as_matrix_gensim(data["Title"].values, max_len)
    batch["FullDescription"] = as_matrix_gensim(data["FullDescription"].values, max_len)
    batch['Categorical'] = categorical_vectorizer.transform(data[categorical_columns].apply(dict, axis=1))
    
    if word_dropout != 0:
        batch["FullDescription"] = apply_word_dropout(batch["FullDescription"], 1. - word_dropout)
    
    if TARGET_COLUMN in data.columns:
        batch[TARGET_COLUMN] = data[TARGET_COLUMN].values
    
    return to_tensors(batch, device)
```

```python
def iterate_minibatches_gensim(data, batch_size=256, shuffle=True, cycle=False, **kwargs):
    """ iterates minibatches of data in random order """
    while True:
        indices = np.arange(len(data))
        if shuffle:
            indices = np.random.permutation(indices)

        for start in range(0, len(indices), batch_size):
            batch = make_batch_gensim(data.iloc[indices[start : start + batch_size]], **kwargs)
            yield batch
        
        if not cycle: break
```

```python
def return_metrics_gensim(model, data, batch_size, name="", **kw):
    squared_error = abs_error = num_samples = 0.0
    model.eval()
    with torch.no_grad():
        for batch in iterate_minibatches_gensim(data, batch_size=batch_size, shuffle=False, **kw):
            batch_pred = model(batch)
            squared_error += torch.mean(torch.square(batch_pred - batch[TARGET_COLUMN]))
            abs_error += torch.mean(torch.abs(batch_pred - batch[TARGET_COLUMN]))
            num_samples += len(batch)
    mse = squared_error.detach().cpu().numpy() / num_samples
    mae = abs_error.detach().cpu().numpy() / num_samples
    return mse, mae
```

```python
def as_matrix_fixed(sequences, max_len):
    """ Convert a list of tokens into a matrix with padding """
    if isinstance(sequences[0], str):
        sequences = list(map(str.split, sequences))
    
    matrix = np.full((len(sequences), max_len), np.int32(PAD_IX))
    for i,seq in enumerate(sequences):
        row_ix = [token_to_id.get(word, UNK_IX) for word in seq[:max_len]]
        matrix[i, :len(row_ix)] = row_ix
    
    return matrix
```

```python
def make_batch_fixed(data, max_len, word_dropout=0, device=torch.device('cpu')):
    """
    Creates a keras-friendly dict from the batch data.
    :param word_dropout: replaces token index with UNK_IX with this probability
    :returns: a dict with {'title' : int64[batch, title_max_len]
    """
    batch = {}
    batch['Id'] = data['Id'].values
    batch["Title"] = as_matrix_fixed(data["Title"].values, max_len)
    batch["FullDescription"] = as_matrix_fixed(data["FullDescription"].values, max_len)
    batch['Categorical'] = categorical_vectorizer.transform(data[categorical_columns].apply(dict, axis=1))
    
    if word_dropout != 0:
        batch["FullDescription"] = apply_word_dropout(batch["FullDescription"], 1. - word_dropout)
    
    if TARGET_COLUMN in data.columns:
        batch[TARGET_COLUMN] = data[TARGET_COLUMN].values
    
    return to_tensors(batch, device)
```

```python
def iterate_minibatches_fixed(data, batch_size=256, max_len=30, shuffle=True, cycle=False, **kwargs):
    """ iterates minibatches of data in random order """
    while True:
        indices = np.arange(len(data))
        if shuffle:
            indices = np.random.permutation(indices)

        for start in range(0, len(indices), batch_size):
            batch = make_batch_fixed(data.iloc[indices[start : start + batch_size]], max_len=max_len, **kwargs)
            yield batch
        
        if not cycle: break
```

```python
def return_metrics_fixed(model, data, batch_size, max_len=30, name="", **kw):
    squared_error = abs_error = num_samples = 0.0
    model.eval()
    with torch.no_grad():
        for batch in iterate_minibatches_fixed(data, batch_size=batch_size, max_len=max_len, shuffle=False, **kw):
            batch_pred = model(batch)
            squared_error += torch.mean(torch.square(batch_pred - batch[TARGET_COLUMN]))
            abs_error += torch.mean(torch.abs(batch_pred - batch[TARGET_COLUMN]))
            num_samples += len(batch)
    mse = squared_error.detach().cpu().numpy() / num_samples
    mae = abs_error.detach().cpu().numpy() / num_samples
    return mse, mae
```

```python
class SalaryPredictorMiddleWordAndGRU(nn.Module):
    def __init__(self, n_tokens=len(tokens), n_cat_features=len(categorical_vectorizer.vocabulary_), hid_size=4, max_len=30):
        super().__init__()
        self.embedder = nn.Embedding(n_tokens, hid_size)
        self.gru = nn.GRU(max_len, hid_size, 2)
        self.title_encoder = nn.Sequential(
            nn.GroupNorm(hid_size, hid_size),
            nn.Dropout(p=0.1),
            nn.ReLU(),
            nn.AdaptiveMaxPool1d(output_size=1)
        )
        self.description_encoder = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.GroupNorm(hid_size, hid_size),
            nn.Dropout(p=0.3),
            nn.ReLU(),
            nn.AdaptiveMaxPool1d(output_size=1)
        )
        self.categorical_encoder = nn.Sequential(
            nn.Linear(n_cat_features, hid_size * 2),
            nn.ReLU(),
            nn.Linear(hid_size * 2, hid_size * 2),
            nn.ReLU()
        )
        self.final_predictor = nn.Sequential(
            nn.Linear(hid_size * 4, hid_size),
            nn.ReLU(),
            nn.Linear(hid_size, 1)
        )


    def pick_middle_words(self, batch):
        middle_words = torch.Tensor([])
        for sentence in batch:
            normalized_sentence = sentence[sentence != PAD_IX]
            middle_word = normalized_sentence[len(normalized_sentence) // 2]
            middle_words = torch.cat([
                                      middle_words, torch.Tensor([middle_word])
                                    ])
        return middle_words.long().unsqueeze(1)

        
    def forward(self, batch):
        title_middle_words = self.pick_middle_words(batch['Title'])
        title_embeddings = self.embedder(title_middle_words).permute(0, 2, 1)
        title_features = self.title_encoder(title_embeddings).squeeze()

        description_embeddings = self.embedder(batch['FullDescription']).permute(0, 2, 1)
        description_features_gru, (h_n, c_n) = self.gru(description_embeddings)
        description_features = self.description_encoder(description_features_gru).squeeze()

        categorical_features = self.categorical_encoder(batch['Categorical'])

        features = torch.cat(
            [title_features, description_features, categorical_features], dim=1)
        
        return self.final_predictor(features).squeeze()
```

```python
# https://github.com/xuyxu/Soft-Decision-Tree

class SDT(nn.Module):
    def __init__(self, input_dim, output_dim, depth=5, lamda=1e-3, use_cuda=False):
        super(SDT, self).__init__()

        self.input_dim = input_dim
        self.output_dim = output_dim

        self.depth = depth
        self.lamda = lamda
        self.device = torch.device("cuda" if use_cuda else "cpu")

        self._validate_parameters()

        self.internal_node_num_ = 2 ** self.depth - 1
        self.leaf_node_num_ = 2 ** self.depth

        self.penalty_list = [
            self.lamda * (2 ** (-depth)) for depth in range(0, self.depth)
        ]

        self.inner_nodes = nn.Sequential(
            nn.Linear(self.input_dim + 1, self.internal_node_num_, bias=False),
            nn.Sigmoid(),
        )

        self.leaf_nodes = nn.Linear(self.leaf_node_num_, self.output_dim, bias=False)


    def forward(self, X, is_training_data=False):
        _mu, _penalty = self._forward(X)
        y_pred = self.leaf_nodes(_mu)
        if is_training_data:
            return y_pred, _penalty
        else:
            return y_pred


    def _forward(self, X):
        batch_size = X.size()[0]
        X = self._data_augment(X)

        path_prob = self.inner_nodes(X)
        path_prob = torch.unsqueeze(path_prob, dim=2)
        path_prob = torch.cat((path_prob, 1 - path_prob), dim=2)

        _mu = X.data.new(batch_size, 1, 1).fill_(1.0)
        _penalty = torch.tensor(0.0).to(self.device)

        begin_idx = 0
        end_idx = 1

        for layer_idx in range(0, self.depth):
            _path_prob = path_prob[:, begin_idx:end_idx, :]

            _penalty = _penalty + self._cal_penalty(layer_idx, _mu, _path_prob)
            _mu = _mu.view(batch_size, -1, 1).repeat(1, 1, 2)

            _mu = _mu * _path_prob

            begin_idx = end_idx
            end_idx = begin_idx + 2 ** (layer_idx + 1)

        mu = _mu.view(batch_size, self.leaf_node_num_)

        return mu, _penalty


    def _cal_penalty(self, layer_idx, _mu, _path_prob):
        penalty = torch.tensor(0.0).to(self.device)

        batch_size = _mu.size()[0]
        _mu = _mu.view(batch_size, 2 ** layer_idx)
        _path_prob = _path_prob.view(batch_size, 2 ** (layer_idx + 1))

        for node in range(0, 2 ** (layer_idx + 1)):
            alpha = torch.sum(
                _path_prob[:, node] * _mu[:, node // 2], dim=0
            ) / torch.sum(_mu[:, node // 2], dim=0)

            coeff = self.penalty_list[layer_idx]

            penalty -= 0.5 * coeff * (torch.log(alpha) + torch.log(1 - alpha))

        return penalty


    def _data_augment(self, X):
        batch_size = X.size()[0]
        X = X.view(batch_size, -1)
        bias = torch.ones(batch_size, 1).to(self.device)
        X = torch.cat((bias, X), 1)

        return X


    def _validate_parameters(self):

        if not self.depth > 0:
            msg = ("The tree depth should be strictly positive, but got {} instead.")
            raise ValueError(msg.format(self.depth))

        if not self.lamda >= 0:
            msg = ("The coefficient of the regularization term should not be negative, but got {} instead.")
            raise ValueError(msg.format(self.lamda))
```

SoftMaxPooling вместо AdaptiveMaxPooling

$$ out_{i, t} = \sum_t {h_{i,t} \cdot {{e ^ {h_{i, t}}} \over \sum_\tau e ^ {h_{j, \tau}} } }$$

```python
def softmax(x):
    exps = torch.exp(x)  # e^(h_i, t)
    sum_exps = torch.sum(exps, dim=1).unsqueeze(-1)  # sum_T e^(h_j, T)
    ones = torch.ones(sum_exps.shape)
    ones_by_sum_exps = ones.div(sum_exps)  # 1 / (...)
    numer = torch.mul(x, exps)  # h_i, t * e^(h_i, t)
    return torch.Tensor(numer @ ones_by_sum_exps).squeeze()
```

```python
class SoftMaxPooling(nn.Module):
    def __init__(self):
        super().__init__()

    
    def forward(self, x: torch.Tensor):
        exps = torch.exp(x)  # e^(h_i, t)
        sum_exps = torch.sum(exps, dim=1).unsqueeze(-1)  # sum_T e^(h_j, T)
        ones = torch.ones(sum_exps.shape)
        ones_by_sum_exps = ones.div(sum_exps)  # 1 / (...)
        numer = torch.mul(x, exps)  # h_i, t * e^(h_i, t)
        return torch.Tensor(numer @ ones_by_sum_exps).squeeze()
```

```python
class SalaryPredictorSoftMax(nn.Module):
    def __init__(self, n_tokens=len(tokens), n_cat_features=len(categorical_vectorizer.vocabulary_), hid_size=4):
        super().__init__()
        self.embedder = nn.Embedding(n_tokens, hid_size)
        self.title_encoder = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.GroupNorm(hid_size, hid_size),
            nn.Dropout(p=0.1),
            SoftMaxPooling()
        )
        self.description_encoder = nn.Sequential(
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.Conv1d(hid_size, hid_size, kernel_size=1),
            nn.GroupNorm(hid_size, hid_size),
            nn.Dropout(p=0.3),
            SoftMaxPooling()
        )
        self.categorical_encoder = nn.Sequential(
            nn.Linear(n_cat_features, hid_size * 2),
            nn.ReLU(),
            nn.Linear(hid_size * 2, hid_size * 2),
            nn.ReLU()
        )
        self.final_predictor = nn.Sequential(
            nn.Linear(hid_size * 4, hid_size),
            nn.ReLU(),
            nn.Linear(hid_size, 1)
        )

        
    def forward(self, batch):
        title_embeddings = self.embedder(batch['Title']).permute(0, 2, 1)
        title_features = self.title_encoder(title_embeddings).squeeze()

        description_embeddings = self.embedder(batch['FullDescription']).permute(0, 2, 1)
        description_features = self.description_encoder(description_embeddings).squeeze()

        categorical_features = self.categorical_encoder(batch['Categorical'])

        features = torch.cat(
            [title_features, description_features, categorical_features], dim=1)
        
        return self.final_predictor(features).squeeze()
```
