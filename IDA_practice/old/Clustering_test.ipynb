# Clustering

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import scipy
from sklearn.neighbors import KNeighborsClassifier

# чтоб картинки рисовались в тетрадке
%matplotlib inline 
```

```python
data = pd.read_csv('adult_data.csv')
data = data.sample(100, random_state=123)
data.head()
```

```python
avail = [col for col, typ in zip(data.columns, data.dtypes) if '64' in str(typ)]

data = data[avail]
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.DataFrame(MinMaxScaler().fit_transform(data), columns=data.columns)
data
```

```python
db = DBSCAN(eps=0.3, min_samples=3)
db.fit_predict(data)
```

```python
db.components_
```

```python
pd.DataFrame(db.components_)
```

```python
print(db.components_[0])
print(db.components_[1])
```

```python
from scipy.spatial import distance
```

```python
distance.pdist(db.components_)
```

```python
distance.cdist(db.components_, db.components_)
```

```python
distance.cdist(db.components_, db.components_)[0][1]
```

```python
distance.euclidean(db.components_[0], db.components_[1])
```
